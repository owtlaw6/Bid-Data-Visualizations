Correlation Matrix Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data_path = './cancer patient data sets.csv'
cancer_data = pd.read_csv(data_path)

# Calculating the correlation matrix
correlation_matrix = cancer_data.drop(['index', 'Patient Id'], axis=1).corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Sankey Diagram

```python
import plotly.graph_objects as go

# Data for the Sankey diagram
label = ['Low', 'Medium', 'High', 'Alcohol Low', 'Alcohol High', 'Pollution Low', 'Pollution High']
source = [0, 0, 1, 1, 2, 2]  # indices correspond to labels
target = [3, 4, 4, 5, 5, 6]
value = [10, 40, 30, 20, 15, 25]

# Create the diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
      pad=15,
      thickness=20,
      line=dict(color="black", width=0.5),
      label=label
    ),
    link=dict(
      source=source,
      target=target,
      value=value
    ))])

fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.show()
```

Chord Diagram

```python
# Required to display plots in Jupyter Notebooks
%matplotlib inline
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

import pandas as pd

# Data setup
data = {
    'Source': ['Low', 'Low', 'Medium', 'Medium', 'High'],
    'Target': ['Alcohol High', 'Pollution Low', 'Alcohol High', 'Pollution High', 'Alcohol Low'],
    'Value': [10, 20, 15, 25, 30]
}
df = pd.DataFrame(data)

# Creating a Holoviews dataset
edges = hv.Dataset(df, ['Source', 'Target'], 'Value')

# Creating the Chord diagram
chord = hv.Chord(edges).opts(
    opts.Chord(
        cmap='Category20', 
        edge_cmap='Category20', 
        labels='name', 
        edge_color='Source', 
        node_color='index'
    )
)

# Displaying the plot
chord
```
