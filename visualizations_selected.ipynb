Correlation Matrix Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from math import pi

data_path = './cancer patient data sets.csv'
cancer_data = pd.read_csv(data_path)
```

```python
correlation_matrix = cancer_data.drop(['index', 'Patient Id'], axis=1).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Box Plots for Various Variables by Cancer Level

```python
fig, axes = plt.subplots(1, 5, figsize=(18, 6))

sns.boxplot(ax=axes[0], data=cancer_data, x='Level', y='Air Pollution', palette='viridis')
axes[0].set_title('Box Plot of Air Pollution by Cancer Level')
axes[0].set_xlabel('Cancer Level')
axes[0].set_ylabel('Air Pollution')

sns.boxplot(ax=axes[1], data=cancer_data, x='Level', y='Alcohol use', palette='viridis')
axes[1].set_title('Box Plot of Alcohol Use by Cancer Level')
axes[1].set_xlabel('Cancer Level')
axes[1].set_ylabel('Alcohol Use')

sns.boxplot(ax=axes[2], data=cancer_data, x='Level', y='Balanced Diet', palette='viridis')
axes[2].set_title('Box Plot of Balanced Diet by Cancer Level')
axes[2].set_xlabel('Cancer Level')
axes[2].set_ylabel('Balanced Diet')

sns.boxplot(ax=axes[3], data=cancer_data, x='Level', y='Fatigue', palette='viridis')
axes[3].set_title('Box Plot of Fatigue by Cancer Level')
axes[3].set_xlabel('Cancer Level')
axes[3].set_ylabel('Fatigue')

sns.boxplot(ax=axes[4], data=cancer_data, x='Level', y='Coughing of Blood', palette='viridis')
axes[4].set_title('Box Plot of Coughing of Blood by Cancer Level')
axes[4].set_xlabel('Cancer Level')
axes[4].set_ylabel('Coughing of Blood')

plt.tight_layout()
plt.show()
```

Violin Plot for 'Genetic Risk' by Cancer Level

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.violinplot(ax=axes[0], data=cancer_data, x='Level', y='Genetic Risk', palette='viridis')
axes[0].set_title('Violin Plot of Genetic Risk by Cancer Level')
axes[0].set_xlabel('Cancer Level')
axes[0].set_ylabel('Genetic Risk')

sns.violinplot(ax=axes[1], data=cancer_data, x='Level', y='chronic Lung Disease', palette='viridis')
axes[1].set_title('Violin Plot of chronic Lung Disease by Cancer Level')
axes[1].set_xlabel('Cancer Level')
axes[1].set_ylabel('chronic Lung Disease')

plt.show()
```

Doughnut Chart Example

```python
level_counts = cancer_data['Level'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(level_counts, labels=level_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Doughnut Chart of Cancer Levels')
plt.show()
```
